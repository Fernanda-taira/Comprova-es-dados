Agrupamentos

```python
#média de quantidade vendida por categoria
print(df.groupby('Categoria')['Quantidade'].mean())
```

```python
#soma de quantidade vendida por categoria
print(df.groupby('Categoria')['Quantidade'].sum())
```

```python
#multiplas estastísticas por categoria
print(df.groupby('Categoria')['Quantidade'].agg(['mean', 'sum', 'count', 'max','min']))
```

Gráficos

```python
# 1. corrigir e converter a coluna Valor Uni.
df['Valor Uni.'] = df['Valor Uni.'].str.replace(',', '.').astype(float)
# 2. converter para float
df['Valor Uni.'] = df['Valor Uni.'].astype(float)
# 3. converter quantidade, se necesário
df['Quantidade'] = df['Quantidade'].astype(float)
# 4. multiplicar
df['Total Vendido'] = df['Quantidade'] * df['Valor Uni.']
# 5. verificar tipo final
print(df['Total Vendido'].dtype)
```

```python
#gráfico simples de barras
import matplotlib.pyplot
df['Categoria'].value_counts().plot(kind='bar')
plt.show()
```

```python
#histograma de distribuição dos valores de venda total vendido
df['Total Vendido'].plot(kind='hist', bins=20)
plt.show()
```

Criar uma coluna nova VIP ou NAO VIP

```python
#criar uma nova coluna STATUS Do Cliente
import numpy as np
df['STATUS']= np.where(df['Total Vendido'] > 1000, 'VIP', 'NAO VIP')
print(df.head)
```

Trabalhando com datas

```python
#garantir que as informações de data estão no formato de dateline
df['Data'] = pd.to_datetime(df['Data'])
```

```python
#extrair as informações de data
df['Ano'] = df['Data'].dt.year
df['Mes'] = df['Data'].dt.month
df['Dia'] = df['Data'].dt.day
df['Dia da Semana'] = df['Data'].dt.dayofweek
```

Análise de oportunidades

```python
#qual o total de vendas?
print(df['Total Vendido'].sum())
```

```python
#qual a média de vendas?
print(df['Total Vendido'].mean())
```

```python
#qual categoria vende mais?
print(df.groupby('Categoria')['Total Vendido'].sum())
```

```python
#qual é a maior venda individual?
print(df['Total Vendido'].max())
```
